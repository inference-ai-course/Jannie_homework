```python
import json
import pickle
import numpy as np

# Load from JSONL and convert to PKL
data = []
with open("../data/chunks.jsonl", "r") as f:
    for line in f:
        data.append(json.loads(line))

# Save as PKL
with open("../data/chunks.pkl", "wb") as f:
    pickle.dump(data, f)

print(f"Converted {len(data)} chunks from JSONL to PKL format")

# Now you can load from PKL as your original code expects
with open("../data/chunks.pkl", "rb") as f:
    data = pickle.load(f)

embeddings = np.array([d["embedding"] for d in data]).astype("float32")
print(f"Loaded {len(embeddings)} embeddings from PKL file")
```
